```julia
using PowerSystems
using PowerSystemCaseBuilder
using Graphs
using GraphPlot

sys = build_system(PSITestSystems, "c_sys5_uc"; force_build=true)
# sys = build_system(MatpowerTestSystems, "matpower_ACTIVSg2000_sys"; force_build=true)

# numeric indices

# Step 1: Collect all buses, and find the one named "nodeC"
buses = collect(get_components(PowerSystems.Bus, sys))
bus_c = first(filter(b -> get_name(b) == "nodeC", buses))

# Step 2: Collect all Lines and find those connected to nodeC
lines = collect(get_components(Line, sys))
lines_to_remove = filter(line -> begin
    arc = get_arc(line)
    get_name(get_from(arc)) == "nodeC" || get_name(get_to(arc)) == "nodeC"
end, lines)

# Step 3: Remove each of these lines from the system
for line in lines_to_remove
    remove_component!(sys, line)
end



# find_islands(sys)

buses = collect(get_components(Bus, sys))
num_buses = length(buses)

lines = collect(get_components(Line, sys))
num_buses = length(buses)

print("lines", lines)

# Initialize undirected graph
g = SimpleGraph(num_buses)




# Create a mapping from bus names to indices
bus_index = Dict(b => i for (i, b) in enumerate(buses))

# Add edges from lines
for line in lines
    arc = get_arc(line)
    add_edge!(g, bus_index[get_from(arc)], bus_index[get_to(arc)])
end




plot = gplot(g)
islands = connected_components(g)
display(plot)

show(g)

print("\nconnected_components", islands)
```
