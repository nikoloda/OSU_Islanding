```julia
using PowerModels
using PowerPlots
using Graphs

# sys = PowerModels.parse_file("case2383wp.m")
sys = PowerModels.parse_file("case14.m")

# Create a graph where buses are nodes and branches are edges
# The internal graph 'case_PMG.graph' uses sequential integer IDs (1 to n) for vertices
case_PMG = PowerModelsGraph(sys, node_components=["bus"], edge_components=["branch"],);

g = Graphs.SimpleGraph(case_PMG.graph) # Does the graph contain cycles?
is_cyclic(g)

# Get the adjacency matrix
# adjacency_matrix(case_PMG.graph) # You can keep this line or remove it as needed

branch_keys = collect(keys(sys["branch"]))

# # Loop through all branches
for branch_id in branch_keys
    # Extract current branch
    branch = sys["branch"][branch_id]

    # Departure bus ID (external ID from the .m file)
    f_id_ext = string(branch["f_bus"])
    # Destination bus ID (external ID from the .m file)
    t_id_ext = string(branch["t_bus"])

    # Departure zone (1-6)
    f_zone = sys["bus"][f_id_ext]["zone"]
    # Destination zone (1-6)
    t_zone = sys["bus"][t_id_ext]["zone"]
        
    if f_zone == t_zone
        continue
    end
        
    # If zones differ, we need to check if zone 6, if not, delete the branch
    if f_zone != t_zone
        
        # We need to find the internal vertex ID used by the graph 'g'
        # PowerModelsGraph stores a map from external IDs to internal sequential IDs (1 to n)
        f_id_int = case_PMG.bus_map[parse(Int, f_id_ext)]
        t_id_int = case_PMG.bus_map[parse(Int, t_id_ext)]
        
        # Dynamically calculate the degree using the Graphs.jl function 'degree(graph, vertex_id)'
        f_deg = degree(g, f_id_int)
        t_deg = degree(g, t_id_int)

        # Case 1: From bus is zone 6  adopt zone from t_bus
        if f_zone == 6
            sys["bus"][f_id_ext]["zone"] = t_zone
            continue  # now zones match, so don't delete

        # Case 2: To bus is zone 6  adopt zone from f_bus
        elseif t_zone == 6
            sys["bus"][t_id_ext]["zone"] = f_zone
            continue
        
        # If f_bus is a leaf (degree=1), adopt t_zone
        elseif f_deg == 1
            println("LEAF branch:", branch_id, "  from:", f_id_ext, "  to:", t_id_ext, "   f zone:", sys["bus"][f_id_ext]["zone"], "   t_zone:", sys["bus"][t_id_ext]["zone"])
            sys["bus"][f_id_ext]["zone"] = t_zone
            continue
        
        # If t_bus is a leaf (degree=1), adopt f_zone
        elseif t_deg == 1
            println("LEAF branch:", branch_id, "  from:", f_id_ext, "  to:", t_id_ext, "   f zone:", sys["bus"][f_id_ext]["zone"], "   t_zone:", sys["bus"][t_id_ext]["zone"])

            sys["bus"][t_id_ext]["zone"] = f_zone
            continue
        
        else
        # Otherwise: both are non-6 and different, and neither is a leaf  delete branch

        println("Disconnecting branch:", branch_id, "  from:", f_id_ext, "  to:", t_id_ext, "   f zone:", sys["bus"][f_id_ext]["zone"], "   t_zone:", sys["bus"][t_id_ext]["zone"])
            
        delete!(sys["branch"], branch_id)
        end
    end
end
```
